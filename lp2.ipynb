```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df = pd.read_csv('learnings.csv')
df.head()
```

```python
df['description'] = df['description'].fillna(df['description'].mode()[0]) 
```

```python
df.drop_duplicates(subset=['title'],inplace=True)
```

```python
df.shape
```

```python
df['skills_list'] = df['skills'].apply(lambda x: [s.strip() for s in x.split(',')])
```

```python
all_skills = [skill for sublist in df['skills_list'] for skill in sublist]
```

```python
skill_count = Counter(all_skills)
print('Top 20 skills:\n')
print(skill_count.most_common(20))
```

```python
num_courses = df.shape[0]
print(f"\nTotal number of courses: {num_courses}")

spc = df['skills'].apply(len)
print(f"\nAverage skills per course: {spc}")
```
